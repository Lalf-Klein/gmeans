```python
from gmeans import GMeans
from sklearn.datasets import make_blobs
import numpy as np
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
X, label = make_blobs(n_samples=500, n_features=2, centers=5, random_state=1, cluster_std=1)
```

```python
gm = GMeans(p_criteria=0.0001)
```

```python
gm.fit(X)
```

```python
labels=gm.fit_predict(X)
```

```python
np.unique(labels)
```

```python
plot_data = pd.DataFrame(X, columns=["x1", "x2"])
```

```python
plot_data["label"] = labels
```

```python
sns.lmplot(x="x1", y="x2", data=plot_data, hue='label', fit_reg=False)
plt.show()
```
